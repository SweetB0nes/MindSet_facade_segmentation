  Искусственный интеллект (ИИ) становится все более популярным во многих отраслях, включая строительство, проектирование, эксплуатацию и продажу квартир.

  1. Проектирование зданий и инфраструктуры:
ИИ может быть использован для проектирования зданий и инфраструктуры, позволяя дизайнерам и инженерам быстро создавать и оценивать различные варианты проектов. Например, ИИ может использоваться для анализа местности, планирования транспортных маршрутов и определения оптимальной расстановки зданий и объектов.
  2. Мониторинг строительства:
ИИ может использоваться для мониторинга строительства, позволяя инженерам и менеджерам отслеживать прогресс работ, определять проблемы и улучшать эффективность процесса. Например, ИИ может использоваться для анализа изображений с камер видеонаблюдения на строительной площадке, чтобы определить, сколько материалов было использовано и сколько еще нужно.
  3. Предсказание срока и стоимости строительства:
ИИ может быть использован для прогнозирования срока и стоимости строительства на основе данных о прошлых проектах, местности и других факторов. Например, ИИ может использоваться для определения оптимального времени и стоимости строительства в зависимости от времени года, рыночных условий и других факторов.
  4. Управление энергопотреблением:
ИИ может использоваться для управления энергопотреблением в зданиях, позволяя оптимизировать системы отопления, вентиляции и кондиционирования воздуха, освещения и других энергоэффективных систем. Например, ИИ может использоваться для анализа погоды, чтобы определить оптимальные настройки систем отопления и кондиционирования воздуха.
  5. Оценка недвижимости:
ИИ может быть использован для оценки недвижимости на основе различных факторов, таких как расположение, площадь, количество комнат, вид на море и т.д. Например, ИИ может использоваться для ана
  6. Планирование обслуживания и ремонта:
ИИ может использоваться для планирования обслуживания и ремонта зданий и инфраструктуры, позволяя эффективно распределять ресурсы и предотвращать возможные поломки. Например, ИИ может использоваться для анализа данных о состоянии оборудования и определения оптимального времени для его замены или ремонта.
  7. Управление безопасностью:
ИИ может использоваться для управления безопасностью на строительной площадке и в зданиях. Например, ИИ может использоваться для мониторинга обнаружения несанкционированного доступа, пожарной безопасности, определения зон повышенного риска и т.д.
  8. Анализ потока людей:
ИИ может использоваться для анализа потока людей в зданиях и на улицах, позволяя улучшить процессы безопасности, обслуживания и планирования. Например, ИИ может использоваться для анализа видеокамер и определения наиболее популярных маршрутов движения людей, оптимизации процессов проверки безопасности и т.д.
  9. Управление операциями продажи и аренды:
ИИ может использоваться для управления операциями продажи и аренды недвижимости, позволяя более эффективно и точно оценивать потребности клиентов, управлять инвентарем и определять оптимальные цены. Например, ИИ может использоваться для анализа данных о покупателях и предоставления персонализированной рекомендации объектов недвижимости.
  10. Мониторинг качества воздуха и воды:
ИИ может использоваться для мониторинга качества воздуха и воды в зданиях и на территории строительной площадки. Например, ИИ может использоваться для анализа данных о составе воздуха и воды, определения оптимальных параметров для их качественной очистки и обеспечения безопасности работников и жителей в зданиях.

```python
class Dataset(torch.utils.data.Dataset):
    
      def __init__(
          self,
          transform = None,
          classes = None,
          augmentation = None
          ):
        self.imgpath_list = sorted(glob('/content/gdrive/MyDrive/facade_datset/images1/*.png'))
        self.labelpath_list = sorted(glob('/content/gdrive/MyDrive/facade_datset/mask2/*.png'))

      def __getitem__(self, i):
                
        imgpath = self.imgpath_list[i]
        img = cv2.imread(imgpath)
        img = cv2.resize(img, dsize = (256, 256))
        # img = img.resize((256, 256), resample=Image.NEAREST)
        img = img / 255
        img = torch.from_numpy(img.astype(np.float32)).clone()
        img = img.permute(2, 0, 1)
        # img = img.permute(1, 2, 0)

        labelpath = self.labelpath_list[i]
        label = Image.open(labelpath)
        label = np.asarray(label)
        label = cv2.resize(label, dsize = (256, 256))
        # label = label.resize((256, 256), resample=Image.BICUBIC)
        label = torch.from_numpy(label.astype(np.float32)).clone()
  
        label = torch.nn.functional.one_hot(label.long(), num_classes = len(class_list))
        label = label.to(torch.float32)
        label = label.permute(2, 0, 1)
        # label = label.permute(1, 2, 0)

        data = {"img": img, "label": label}
        return data

      def __len__(self):
            return len(self.imgpath_list)
```

U-net


```python
class TwoConvBlock(nn.Module):
    def __init__(self, in_channels, middle_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels, middle_channels, kernel_size = 3, padding = "same")
        self.bn1 = nn.BatchNorm2d(middle_channels)
        self.rl = nn.ReLU()
        self.conv2 = nn.Conv2d(middle_channels, out_channels, kernel_size = 3, padding = "same")
        self.bn2 = nn.BatchNorm2d(out_channels)

    def forward(self, x):
        x = self.conv1(x)
        x = self.bn1(x)
        x = self.rl(x)
        x = self.conv2(x)
        x = self.bn2(x)
        x = self.rl(x)
        return x

class UpConv(nn.Module):
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.up = nn.Upsample(scale_factor = 2, mode = "bilinear", align_corners = True)
        self.bn1 = nn.BatchNorm2d(in_channels)
        self.conv = nn.Conv2d(in_channels, out_channels, kernel_size = 2, padding = "same")
        self.bn2 = nn.BatchNorm2d(out_channels)

    def forward(self, x):
        x = self.up(x)
        x = self.bn1(x)
        x = self.conv(x)
        x = self.bn2(x)
        return x

class UNet_2D(nn.Module):
    def __init__(self):
        super().__init__()
        self.TCB1 = TwoConvBlock(3, 64, 64)
        self.TCB2 = TwoConvBlock(64, 128, 128)
        self.TCB3 = TwoConvBlock(128, 256, 256)
        self.TCB4 = TwoConvBlock(256, 512, 512)
        self.TCB5 = TwoConvBlock(512, 1024, 1024)
        self.TCB6 = TwoConvBlock(1024, 512, 512)
        self.TCB7 = TwoConvBlock(512, 256, 256)
        self.TCB8 = TwoConvBlock(256, 128, 128)
        self.TCB9 = TwoConvBlock(128, 64, 64)
        self.maxpool = nn.MaxPool2d(2, stride = 2)
        
        self.UC1 = UpConv(1024, 512) 
        self.UC2 = UpConv(512, 256) 
        self.UC3 = UpConv(256, 128) 
        self.UC4= UpConv(128, 64)

        self.conv1 = nn.Conv2d(64, 4, kernel_size = 1)
        self.soft = nn.Softmax(dim = 1)

    def forward(self, x):
        x = self.TCB1(x)
        x1 = x
        x = self.maxpool(x)

        x = self.TCB2(x)
        x2 = x
        x = self.maxpool(x)

        x = self.TCB3(x)
        x3 = x
        x = self.maxpool(x)

        x = self.TCB4(x)
        x4 = x
        x = self.maxpool(x)

        x = self.TCB5(x)

        x = self.UC1(x)
        x = torch.cat([x4, x], dim = 1)
        x = self.TCB6(x)

        x = self.UC2(x)
        x = torch.cat([x3, x], dim = 1)
        x = self.TCB7(x)

        x = self.UC3(x)
        x = torch.cat([x2, x], dim = 1)
        x = self.TCB8(x)

        x = self.UC4(x)
        x = torch.cat([x1, x], dim = 1)
        x = self.TCB9(x)

        x = self.conv1(x)

        return x
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
unet = UNet_2D().to(device)
optimizer = optim.Adam(unet.parameters(), lr = 0.001)
```

```python
TverskyLoss = smp.losses.TverskyLoss(mode = 'multilabel', log_loss = False)
BCELoss     = smp.losses.SoftBCEWithLogitsLoss()
def criterion(pred,target):
    return 0.5 * BCELoss(pred, target) + 0.5 * TverskyLoss(pred, target)
```

Обучение, проверка, 
подсчет окон


```python
history = {"train_loss": []}
n = 0
m = 0
val_loss_min = np.Inf
validation_accuracy = []
validation_loss = []

for epoch in range(30):
  train_accuracy = 0
  train_loss = 0
  val_accuracy = 0
  val_loss = 0
  correct_pixels = 0
  total_pixels = 0

# training
  unet.train()
  for i, data in enumerate(train_loader):
    inputs, labels = data["img"].to(device), data["label"].to(device)
    optimizer.zero_grad()
    outputs = unet(inputs)
    
    # pixel accuracy
    thresholded_outputs = (outputs > 0.5).int()
    correct_pixels = (thresholded_outputs == labels).sum().item()
    total_pixels += (inputs.size(2) * inputs.size(3)) * inputs.size(0)
    train_acc = correct_pixels / total_pixels
    train_accuracy += train_acc
    
    # loss
    loss = criterion(outputs, labels)
    loss.backward()
    optimizer.step()
    train_loss += loss.item()
    history["train_loss"].append(loss.item())
    n += 1
    if i % ((len(train) // batch_size) // 10) == (len(train) // batch_size) // 10 - 1:
      print(f"epoch:{epoch + 1}  index:{i + 1}  train_accuracy:{train_accuracy / n:.5f}")
      print(f"epoch:{epoch + 1}  index:{i + 1}  train_loss:{train_loss / n:.5f}")
      n = 0
      train_accuracy = 0
      train_loss = 0

# validation
  unet.eval()
  with torch.no_grad():
    for i, data in enumerate(val_loader):
      inputs, labels = data["img"].to(device), data["label"].to(device)
      outputs = unet(inputs)
       
      # pixel accuracy
      thresholded_outputs = (outputs > 0.5).int()
      correct_pixels = (thresholded_outputs == labels).sum().item()
      total_pixels += (inputs.size(2) * inputs.size(3)) * inputs.size(0)
      val_acc = correct_pixels / total_pixels
      val_accuracy += val_acc  
       
      # loss
      loss = criterion(outputs, labels)
      val_loss += loss.item()
      m += 1
      if i % (len(val) // batch_size) == len(val) // batch_size - 1:
        print(f"epoch:{epoch + 1}  index:{i + 1}  validation_accuracy:{val_accuracy / m:.5f}")
        print(f"epoch:{epoch + 1}  index:{i + 1}  validation_loss:{val_loss / m:.5f}")
        validation_accuracy.append(val_accuracy)
        validation_loss.append(val_loss)
        if val_loss < val_loss_min:
            val_loss_min = val_loss
            torch.save(unet.state_dict(), 'semantic_segmentation.pt')
            print('Detected network improvement, saving current model')
        
        m = 0
        val_accuracy = 0
        val_loss = 0
        
print("finish training")
```
